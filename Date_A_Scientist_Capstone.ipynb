```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Investigating the data
profiles = pd.read_csv('profiles.csv')
print(profiles.columns)
profiles.head()
```

```python
# Fixing column values with regex
profiles = profiles.replace(to_replace ='&rsquo;', value = '\'', regex = True) 
profiles.head()
```

```python
# Investigating the age range of OKCupid users
plt.hist(profiles.age, bins = 20)
plt.xlabel('Age Range')
plt.ylabel('Frequency')
plt.xlim(16, 80)
plt.show()
```

```python
# Observing the drinking, smoking, and drug use habits of users
print('DRINKS:')
print(profiles.drinks.value_counts())
print('SMOKES:')
print(profiles.smokes.value_counts())
print('DRUGS:')
print(profiles.drugs.value_counts())
```

```python
# Mapping categorical data
profiles['drinks'] = profiles['drinks'].map({'not at all': 0, 'rarely': 1, 'socially': 2, 'often': 3, 
                                             'very often': 4, 'desperately': 5})
profiles['smokes'] = profiles['smokes'].map({'no': 0, 'trying to quit': 1, 'when drinking': 2, 'sometimes': 3, 
                                             'yes': 4})
profiles['drugs'] = profiles['drugs'].map({'never': 0, 'sometimes': 1, 'often': 2})

# Removing the NaNs (assuming NaN means user never does said activity)
profiles[['drinks', 'smokes', 'drugs']] = profiles[['drinks', 'smokes', 'drugs']].replace(np.nan, 0, regex=True)
```

```python
profiles.head()
```

```python
# Collecting all the essay responses
essay_cols = ["essay0","essay1","essay2","essay3","essay4","essay5","essay6","essay7","essay8","essay9"]

# Removing the NaNs
all_essays = profiles[essay_cols].replace(np.nan, '', regex=True)
# Combining the essays
all_essays = all_essays[essay_cols].apply(lambda x: ' '.join(x), axis=1)
profiles["essay_length"] = all_essays.apply(lambda x: len(x))
```

```python
profiles.head()
```

```python
# Creating a list of the words in each essay
essay_word_list = []
for essay in all_essays:
    essay_word_list.append(essay.split())
```

```python
essay_word_list[3]
```

```python
# Adding a new feature column: the number of times a user said 'I', or 'me' in their essay responses
i_or_me_total = []
for essay in essay_word_list:
    i_or_me_counter = 0
    for word in essay:
        if word == 'i':
            i_or_me_counter += 1
        elif word == 'me':
            i_or_me_counter += 1
    i_or_me_total.append(i_or_me_counter)
```

```python
# Checking to make sure there is a counter for each essay
print(len(essay_word_list))
print(len(i_or_me_total))
```

```python
# Adding column into dataframe
profiles['i_or_me_count'] = i_or_me_total
profiles.head()
```

# Linear Regression Model

Research question: Can age be predicted by the frequency of "I" or "me" in OKCupid user essay responses? 

```python
from sklearn.linear_model import LinearRegression

X = profiles['i_or_me_count']
X = X.values.reshape(-1, 1)
y = profiles['age']

lr_model = LinearRegression()
lr_model.fit(X, y)
```

```python
print(lr_model.coef_, lr_model.intercept_)

y_predict = lr_model.predict(X)

plt.scatter(X, y)
plt.plot(X, y_predict)
plt.show()
```

Clearly there is not a strong enough relationship bewteen age and the frequency of 'i' or 'me' appearing in the essay responses to generate a useful linear regression model.

# Classification Model #1: K-Nearest Neighbors

Research question: Can a classifier be built to predict a user's self-described body type based on their drinking, smoking, and drug use habits? 

```python
# Checking for and removing NaNs
profiles = profiles[profiles['body_type'].notna()]
profiles['body_type'].value_counts()
```

```python
# Normalizing the data
from sklearn.preprocessing import scale, LabelEncoder

# Establishing scaled labels
labels = profiles['body_type']
encoder = LabelEncoder()
scaled_labels = encoder.fit_transform(labels)
body_type_mappings= {index: label for index, label in 
                  enumerate(encoder.classes_)}
print(body_type_mappings)
profiles['scaled_body_type'] = scaled_labels

# Establishing scaled features
features = profiles[['drinks', 'drugs', 'smokes']]
scaled_features = scale(features, axis=0)

print(scaled_features[0])
profiles.head()
```

# KNN

```python
# Train test split
from sklearn.model_selection import train_test_split

train_features, test_features, train_labels, test_labels = train_test_split(scaled_features, scaled_labels, 
                                                                            test_size = 0.2, random_state = 1)
```

```python
# Building and fitting the KNN classifier
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors = 100)
classifier.fit(train_features, train_labels)
print(classifier.score(test_features, test_labels))
```

```python
# Finding the best value for k
scores = []
for k in range(75, 150):
    KNN = KNeighborsClassifier(n_neighbors = k)
    KNN.fit(train_features, train_labels)
    scores.append(KNN.score(test_features, test_labels))
plt.plot(range(75, 150), scores)
plt.xlabel('K')
plt.ylabel('Classifier Score')
plt.show()

# returning the best score and accompanying value of k
k = range(75, 150)
best_score = max(scores)
best_k = k[scores.index(best_score)]
print("The best classifier score is " + str(best_score) + " where k = " + str(best_k))
```

The accuracy we would expect from predicting a user's body type by randomly selecting one would be 1/12, or 0.0833. Although this model is about 3 times more accurate, a score of 0.27 is still a poor score for a machine learning model. Therefore, this model is unable to show a significant relationship between drinking, smoking, and drug habits and an OKCupid user's self proclaimed body type. 

# Classification Model #2: Support Vector Machines

Research Question: Can sex be predicted based on essay length and the frequency of 'I' or 'me' in essay responses?

```python
# checking for NaNs
print(profiles.sex.isna().value_counts())
print(profiles.essay_length.isna().value_counts())
print(profiles.i_or_me_count.isna().value_counts())
profiles[['sex', 'essay_length', 'i_or_me_count']]
```

```python
profiles['sex_mapped'] = profiles['sex'].map({'m': 0, 'f': 1})
print(profiles['sex_mapped'].value_counts())
print(profiles['essay_length'].describe())
print(profiles['i_or_me_count'].describe())
```

```python
plt.scatter(profiles.i_or_me_count, profiles.essay_length, c = profiles.sex_mapped)
plt.xlabel('I or me count')
plt.ylabel('Essay Length')
plt.show()
```

```python
# Normalizing the data
from sklearn.preprocessing import scale

# Establishing labels
svm_labels = profiles['sex_mapped']

# Establishing scaled features
features = profiles[['i_or_me_count', 'essay_length']]
scaled_features = scale(features, axis=0)

print(scaled_features[0])
```

```python
from sklearn.svm import SVC

training_points, test_points, training_labels, test_labels = train_test_split(
scaled_features, svm_labels, train_size = 0.8, test_size = 0.2, random_state = 100)
```

```python
SVMclassifier = SVC(gamma = 10, C = 1)
SVMclassifier.fit(training_points, training_labels)

print(SVMclassifier.score(test_points, test_labels))
```

This SVM model makes accurate classifications more than half of the time, but it still is inaccurate too often to be considered a useful model.
